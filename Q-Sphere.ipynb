```python
from qiskit import QuantumCircuit
import numpy as np
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere
```

```python
n = 2
qc = QuantumCircuit(n)
qc.h(0)
qc.cx(0,1)
statevect = Statevector.from_instruction(qc).data
print(statevect)
plot_state_qsphere(statevect)
```
